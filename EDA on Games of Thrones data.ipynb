# importing modules

```python
import pandas as pd
import numpy as np
```

# loading the data

```python
games = pd.read_csv('battles.csv')
```

```python
games.head()
```

```python
games.shape
```

# Rename

```python
games.rename(columns={'attacker_1': 'primary_attacker'}, inplace=True)
games.head()
```

```python
games.rename(columns={'defender_1': 'primary_defender'}, inplace=True)
games.head()
```

```python
games['attacker_king'].value_counts() #no of times king attacked
```

```python
games['location'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king', y='attacker_size', data = games)
plt.show()
```

In the graph 

1.Joffery/Tommen Baratheon have size around 1000

2.Robb stark have less than 1000

3.Balon/Euron Greyjoy has very less size as compared to other three

4.Stannis Baratheon has the highest size around 20000-30000

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king', y='defender_size', data = games)
plt.show()
```

Here we can see,

1.Renly baratheon has highest size of defender 

2.Joffery/Tommen Baratheon have around 7500-10000

3.Balon/Euron Greyjoy and Mance Rayder has very low defender size

4.Stannis Baratheon is the lowest among all where as it had highest attacker size in above graph

```python
sns.countplot(x= games['attacker_king'], hue=games['battle_type'])
plt.show()
```
